```python
mapping = {
    '$101-200': 3,
    '$201-400': 4,
    '$51-100': 2,
    '$401-800': 5,
    '$50以下': 1
}
```

```python
mapping['$60以下']
```

```python
a = '$60以下'
mapping.get(a, 'not found')
```

```python
def cleaning(e):
    return mapping.get(e, 0)

cleaning('$50以下')
```

```python
df.head()
```

```python
df['price_num'] = df['price'].apply(cleaning)
```
