## Two variables

Is price related with food type?

```python
pd.pivot_table(df, index=['type'], columns=['price'], aggfunc=np.mean)
```

```python
def myfunc(df):
    return len(df)
pd.pivot_table(df, index=['type'], columns=['price'], aggfunc=myfunc).fillna(0)
```

```python
def aggregate_group(df):
    return len(df)
def select_types(e):
    if e in ['海鮮', '火鍋', '甜品/糖水', '烤肉', '壽司/刺身']:
        return True
    else:
        return False
pt_food = pd.pivot_table(
    df[df['type'].apply(select_types)], index=['type'], columns=['price'], aggfunc=aggregate_group
).fillna(0)[[('likes')]]
#).fillna(0)[[('likes', 	'$101-200')]]
```

```python
sns.heatmap(pt_food)
```
