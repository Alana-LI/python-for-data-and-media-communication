```python
def count_retweeted_number(name):
    def check_name(x):
        return name in str(x).lower()
    return df['text'].apply(check_name).value_counts().get(True, 0)
```

```python
all_text = []
for text in df['text']:
    all_text.extend(str(text).split())
```

```python
len(all_text)
```

```python
word_count = {}
```

```python
for word in all_text:
    if word not in word_count:
        word_count[word] = 1
    else:
        word_count[word] = word_count[word] + 1
```

```python
word_count
```

```python
pd.Series(word_count).to_frame().reset_index()
```

```python
pd.Series(word_count).sort_values(ascending=False)
```
