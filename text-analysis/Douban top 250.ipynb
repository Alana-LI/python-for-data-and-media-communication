## How to solve the "2004(中国大陆)     1" entry?

```python
def cleaning(x):
    if x in ['2004(中国大陆)']:
        return False
    else:
        return True
s_year = df[df['year'].apply(cleaning)]['year'].value_counts()
s_year
```

```python
from matplotlib import pyplot as plt
```
