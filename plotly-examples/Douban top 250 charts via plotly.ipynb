**NOTE:**

The above two figures are wrong because it used incorrect data type. When non-`int` type is plotted, they are uniformly located on the x-asis. We need to do conversion to get the right chart. Read more about the data type here: http://chris.friedline.net/2015-12-15-rutgers/lessons/python2/03-data-types-and-format.html

```python
df_hk_cn['index'].dtype
```

```python
df_hk_cn['index-int'] = df_hk_cn['index'].astype('int')
```

```python
df_hk_cn['index-int'].dtype
```

```python
df_hk_cn.set_index('index-int').fillna(0).cumsum().plot(kind='area', stacked=False)
```
