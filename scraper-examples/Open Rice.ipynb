## Open Rice



```python
url = 'https://www.openrice.com/en/hongkong/restaurants?what=sushi'
```

```python
import requests
```

```python
r = requests.get(url)
```

```python
r = requests.get(url, headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.167 Safari/537.36'})
```

```python
r.content
```

```python
import bs4
```

```python
mypage = bs4.BeautifulSoup(r.content)
```

```python
mypage
```

```python
mypage.find_all('h2', attrs={'class': 'title-name'})
```

```python
for myh2 in mypage.find_all('h2', attrs={'class': 'title-name'}):
    title = myh2.text.strip()
    print(title)
```
