```python
from selenium import webdriver
import time
browser = webdriver.Chrome('/users/xuyucan/chromedriver')
```

```python
articles = []
browser.get('http://money.cnn.com/search/index.html?sortBy=date&primaryType=mixed&search=Search&query=trade%20war')
time.sleep(2)

for i in range (0,10):

    browser.find_elements_by_xpath("//div[@id='summaryList_mixed']//div[@class='summaryBlock']")
    for session in browser.find_elements_by_xpath("//div[@id='summaryList_mixed']//div[@class='summaryBlock']"):
        article = {}
        headline = session.find_element_by_class_name("cnnHeadline").text
        article['headline'] = headline
        article['date'] = session.find_element_by_class_name("cnnDateStamp").text
#         url = session.find_element_by_xpath("//div[@class='cnnHeadline']/*[@href]")  #can not directly get the href attributes
        url = session.find_element_by_tag_name("a")
        article['url'] = url.get_attribute('href')
        articles.append(article)
    
    browser.execute_script('window.scrollTo(0, document.body.scrollHeight/1.5);') #try 2: scroll whole body down to the bottom... error: some page's navigation bar has blocked the click button. 
    # try 3: (document.body.scrollHeight - int) ...  fail: can be divided, but can not be minus.
    next_page = browser.find_element_by_link_text('Next') #try 1: just click link by default ... error: not clickable
    print(next_page.get_attribute('href'))
    print(type(next_page.get_attribute('href')))    
    next_page.click()
    i = i + 1
    print(i)
```

```python
import pandas as pd
pd.DataFrame(articles)
```
