This notebook tries to scrape the popup menu of openrice in order to get all the geo regions.

- This shows CSS Selector can be used where xpath is used
- Sometimes xpath is simpler but can be worked around

```python
import selenium
from selenium import webdriver
driver = webdriver.Chrome()
driver.get('https://www.openrice.com/zh/hongkong/restaurants')
```

```python
box = driver.find_element_by_css_selector('.header-searchbar-adv-search-btn')
```

```python
type(box)
```

```python
box.click()
```

```python
district_box = driver.find_element_by_xpath("//div[@class='modal-body or-scrollbar']//section[1]")
```

```python
search_box = district_box.find_element_by_xpath("//div[@data-href='#search/landmark/district']").click()
```
